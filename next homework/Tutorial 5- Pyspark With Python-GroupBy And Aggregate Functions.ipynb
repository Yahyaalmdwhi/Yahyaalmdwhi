### Pyspark GroupBy And Aggregate Functions 

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('YER').getOrCreate()
```

```python
spark
```

```python
df_pyspark=spark.read.csv('fx_rates_sample.csv',header=True,inferSchema=True)
```

```python
df_pyspark.show()
```

```python
df_pyspark.printSchema()
```

```python
## Groupby
### Grouped to find the maximum salary
df_pyspark.groupBy('YER').sum().show()
```

```python
df_pyspark.groupBy('YER').avg().show()
```

```python
### Groupby Departmernts  which gives maximum salary
df_pyspark.groupBy('YER').sum().show()
```

```python
df_pyspark.groupBy('YER').mean().show()
```

```python
df_pyspark.groupBy('YER').count().show()
```

```python
df_pyspark.agg({'YER':'sum'}).show()
```
