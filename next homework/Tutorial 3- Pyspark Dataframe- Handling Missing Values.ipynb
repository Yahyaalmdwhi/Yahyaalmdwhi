### Pyspark Handling Missing Values
- Dropping Columns
- Dropping Rows
- Various Parameter In Dropping functionalities
- Handling Missing values by Mean, MEdian And Mode

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('Practise').getOrCreate()
```

```python
df_pyspark=spark.read.csv('fx_rates_sample.csv',header=True,inferSchema=True)
```

```python
df_pyspark.printSchema()
```

```python
df_pyspark.show()
```

```python
##drop the columns
df_pyspark.drop('Name').show()
```

```python
df_pyspark.show()
```

```python
df_pyspark.na.drop().show()
```

```python
### any==how
df_pyspark.na.drop(how="any").show()
```

```python
##threshold
df_pyspark.na.drop(how="any",thresh=3).show()
```

```python
##Subset
df_pyspark.na.drop(how="any", subset=['SAR']).show()
```

```python
### Filling the Missing Value
df_pyspark.na.fill('Missing Values',['SAR']).show()
```

```python
df_pyspark.show()
```

```python
df_pyspark.printSchema()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['age', 'Experience', 'Salary'], 
    outputCols=["{}_imputed".format(c) for c in ['age', 'Experience', 'Salary']]
    ).setStrategy("median")
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=["SAR"],
    outputCols=["SAR_imputed"]
)

imputed_df = imputer.fit(df_pyspark).transform(df_pyspark)
imputed_df.show(5)
```
