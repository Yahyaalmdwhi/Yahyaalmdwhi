### Pyspark Dataframes
- Filter Operation
- &,|,==
- ~

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('dataframe').getOrCreate()
```

```python
df_pyspark=spark.read.csv('fx_rates_sample.csv',header=True,inferSchema=True)
df_pyspark.show()
```

### Filter Operations

```python
### Salary of the people less than or equal to 20000
df_pyspark.filter("SAR <= 3.76").show()

```

```python
# فلترة على قيمة SAR أقل من 3.76
df_pyspark.filter("SAR <= 3.76").select(["timestamp", "SAR"]).show()
```

```python
# فلترة على العمود SAR (مثال: كل السطور حيث SAR <= 3.76)
df_pyspark.filter(df_pyspark['SAR'] <= 3.76).show()
```

```python
df_pyspark.filter((df_pyspark['SAR'] <= 3.76) | 
                  (df_pyspark['SAR'] >= 3.74)).show()
```

```python
# كل الصفوف حيث SAR ليس <= 3.75
df_pyspark.filter(~(df_pyspark['SAR'] <= 3.75)).show()
```
