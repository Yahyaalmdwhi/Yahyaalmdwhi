```python
### Adding Columns in data frame
# إضافة عمود جديد يعتمد على قيمة EUR
df_pyspark = df_pyspark.withColumn(
    "EUR_plus_2",
    df_pyspark["EUR"] + 2
)
df_pyspark.show(5)
```

```python
df_pyspark.show()
```

```python
### Drop the columns
df_pyspark=df_pyspark.drop('Experience After 2 year')
```

```python
df_pyspark.show()
```

```python
### Rename the columns
df_pyspark.withColumnRenamed('Name','New Name').show()
```
